```python
import pandas as pd
df = pd.read_csv("csv/Nike_korea.csv")
df
```

```python
df["주소"]
```

```python
addr = []

for address in df["주소"]:
    addr.append(str(address).split())

addr


```

```python
addr2 = []
for i in range(len(addr)):
  if addr[i][0] == '서울': addr[i][0] = '서울특별시'
  elif addr[i][0] == '부산시': addr[i][0] = '부산광역시'
  elif addr[i][0] == '서울시': addr[i][0] = '서울특별시'
  elif addr[i][0] == '인천': addr[i][0] = '인천광역시'
  elif addr[i][0] == '광주': addr[i][0] = '광주광역시'
  elif addr[i][0] == '대전시': addr[i][0] = '대전광역시'
  elif addr[i][0] == '울산시': addr[i][0] = '울산광역시'
  elif addr[i][0] == '세종시': addr[i][0] = '세종특별자치시'
  elif addr[i][0] == '경기': addr[i][0] = '경기도'
  elif addr[i][0] == '제주도': addr[i][0] = '제주특별자치도'
  elif addr[i][0] == '충남': addr[i][0] = '충청남도'
  elif addr[i][0] == '충북': addr[i][0] = '충청북도'
  elif addr[i][0] == '전남': addr[i][0] = '전라남도'
  elif addr[i][0] == '전북': addr[i][0] = '전라북도'
  elif addr[i][0] == '제주': addr[i][0] = '제주특별자치도'
  elif addr[i][0] == '제주시': addr[i][0] = '제주특별자치도'
  
  addr2.append(' '.join(addr[i]))
print("데이터 개수:", len(addr2))
addr2
```

```python
addr2 = pd.DataFrame(addr2, columns =["주소"])
addr2


```

```python
df2 = pd.concat([df, addr2], axis = 1)

df2.head()
```

```python
import folium
```

```python
df3 = pd.read_csv("csv/nike_addr.csv")
df3.head()






```

```python
df_merge = pd.merge(df2, df3, how = "inner", left_index = True, right_index = True)
df_merge.head()
```

```python
# 불필요한 열 제거
df_merge = df_merge.drop(columns=["field1", "field2", "field3", "field4", "_GC_TYPE"])

# 첫 번째 "주소" 열을 "우편번호"로 변경
df_merge.columns = df_merge.columns.str.replace('주소', '우편번호', n=1)

# 두 번째 "주소" 열 삭제
df_merge = df_merge.loc[:, ~df_merge.columns.duplicated()]

# _CLEANADDR을 "주소"로 변경
df_merge = df_merge.rename(columns={"_CLEANADDR": "주소"})

```

```python
df_merge.to_csv("csv/nike_korea_final.csv", index = False)
```
