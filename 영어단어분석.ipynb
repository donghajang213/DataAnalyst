```python
import pandas as pd
import glob
import re
from functools import reduce
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import STOPWORDS, WordCloud
```

```python
all_files = glob.glob("text/myCabinetExcelData*.xls")
```

```python
all_files
```

```python
all_files_data = []
for file in all_files:
    data_frame = pd.read_excel(file)
    all_files_data.append(data_frame)
```

```python
all_files_data[0]
```

```python
all_files_data_concat = pd.concat(all_files_data, axis = 0, ignore_index = True)
all_files_data_concat
```

```python
all_title = all_files_data_concat["제목"]
all_title
```

```python
stopWords = set(stopwords.words("english"))
lemma = WordNetLemmatizer()
words = []

for title in all_title:
    EnWords = re.sub(r"[^a-zA-Z]+", " ", str(title)) 
    EnWordsToken = word_tokenize(EnWords.lower())
    EnWordsTokenStop = [w for w in EnWordsToken if w not in stopWords] # 출력값 for 변수 in 반복가능한 객체 if 조
    EnWordsTokenStopLemma = [lemma.lemmatize(w) for w in EnWordsTokenStop]
    words.append(EnWordsTokenStopLemma)
print(words)
```

```python
words2 = list(reduce(lambda x, y: x+y, words))
print(words2) #작업 내용 확인용 출력
```

```python
count = Counter(words2)
count 
```

```python
word_count = dict()

for tag, counts in count.most_common(50):
    if(len(str(tag)) > 1):
        word_count[tag] = counts
        print("%s : %d" %(tag, counts))
```

```python
del word_count["big"]
del word_count["data"]
```

```python
word_count
```

```python
sorted_Keys = sorted(word_count, key = word_count.get, reverse = True)
sorted_Values = sorted(word_count.values(), reverse = True)
plt.bar(range(len(word_count)), sorted_Values, align = 'center')
plt.xticks(range(len(word_count)), list(sorted_Keys), rotation = 85)
plt.show()
```

```python
all_files_data_concat['doc_count'] = 0
summary_year = all_files_data_concat.groupby('출판일', as_index = False)['doc_count'].count()
summary_year #작업 내용 확인용 출력
```

```python
plt.figure(figsize = (12, 5))
plt.xlabel("year")
plt.ylabel("doc-count")
plt.grid(True)
plt.plot(range(len(summary_year)), summary_year['doc_count'])
plt.xticks(range(len(summary_year)), [text for text in summary_year['출판일']])
plt.show()
```

```python
stopwords = set(STOPWORDS)
wc = WordCloud(background_color = 'black', stopwords = stopwords, width = 800, height = 600)
cloud = wc.generate_from_frequencies(word_count)
plt.figure(figsize = (8,8))
plt.imshow(cloud)
plt.axis('on')
plt.show()
```

```python
stopwords = set(STOPWORDS)
wc = WordCloud(background_color = 'aqua', stopwords = stopwords, width = 800, height = 600)
cloud = wc.generate_from_frequencies(word_count)
plt.figure(figsize = (8,8))
plt.imshow(cloud)
plt.axis('off')
plt.show()
```

```python
import json
import re
from konlpy.tag import Okt
from collections import Counter
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
from wordcloud import WordCloud
```

```python
with open("etnews.kr_facebook_2016-01-01_2018-08-01_4차 산업혁명.json", "r", encoding = "utf8") as f:
    data = json.load(f)
```

```python
data
```

```python
message = ""
for item in data:
    if "message" in item.keys():
        message = message + re.sub(r"[^\w]", " ", item["message"]) + ""
message
```

```python
nlp = Okt()
message_N = nlp.nouns(message)
message_N #작업 내용 확인용 출력
```

```python
count = Counter(message_N)
count
```

```python
word_count = dict()
for tag, counts in count.most_common(200):
    if (len(str(tag)) > 1):
        word_count[tag] = counts
        print("%s : %d" % (tag, counts))
```

```python
font_path = "C:/Windows/fonts/malgun.ttf"
font_name = font_manager.FontProperties(fname = font_path).get_name()
matplotlib.rc('font', family = font_name)
```

```python
plt.figure(figsize = (12, 5))
plt.xlabel('키워드')
plt.ylabel('빈도수')
plt.grid(True)
sorted_Keys = sorted(word_count, key = word_count.get, reverse = True)
sorted_Values = sorted(word_count.values(), reverse = True)
plt.bar(range(len(word_count)), sorted_Values, align = 'center')
plt.xticks(range(len(word_count)), list(sorted_Keys), rotation = 75)
plt.show()
```

```python
wc = WordCloud(font_path, background_color = 'ivory', width = 800, height = 600)
cloud = wc.generate_from_frequencies(word_count)
plt.figure(figsize = (8, 8))
plt.imshow(cloud)
plt.axis('off')
plt.show()
```
